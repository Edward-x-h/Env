```python
import pandas as pd
dt=pd.read_csv('微博清单_全球变暖_前100页.csv')
```

```python
import os
```

```python
# Remove the columns
papers = dt.drop(columns=['页码', '微博id', '微博作者','是否认证','发布时间','发布地点','转发数','评论数','点赞数'], axis=1).sample(100)

# Print out the first rows of papers
papers.head()
```

```python
# Load the regular expression library
import re
# Remove punctuation
papers['微博内容_processed'] = \
papers['微博内容'].map(lambda x: re.sub('[，\。！？ “” 【】 #]', '', x))
papers['微博内容_processed'] = \
papers['微博内容_processed'].map(lambda x: x.lower())
```

```python
import gensim
from gensim.utils import simple_preprocess
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words.extend(['from', 'subject', 're', 'edu', 'use'])
def sent_to_words(sentences):
    for sentence in sentences:
        # deacc=True removes punctuations
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) 
             if word not in stop_words] for doc in texts]
data = papers.微博内容_processed.values.tolist()
data_words = list(sent_to_words(data))
# remove stop words
data_words = remove_stopwords(data_words)
print(data_words[:1][0][:30])
```

```python
import gensim.corpora as corpora
# Create Dictionary
id2word = corpora.Dictionary(data_words)
# Create Corpus
texts = data_words
# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]
# View
print(corpus[:1][0][:30])
```

```python
from pprint import pprint
# number of topics
num_topics = 10
# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_topics)
# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```
